## List comprehension.

 Its a way of constructing a list

```python
a_list = [1, 2, 3, 4, 5]
a_squared_list = [i*i for i in a_list]
a_squared_list
```

Python has functions like `enumerate` and `zip`. They allow us to work with list and have some handy functionalities. `enumerate` gives a list of tuples with each tuple of the form `(index, value)`, while the `zip` takes elements from each list and outs them together into a tuple, thus creating a list of tuples. 

```python
enumerate(a_squared_list), zip(a_list, a_squared_list)
```

Pytohn is lazy, and will always return an iterator. To get the values one need either create a `for` loop or use any other operation that works with an iterator.

```python
for i, value in enumerate(a_squared_list):
    print(f'the values of the {i}-th record in the list is {value}')
```

```python
#enumerate is similar to
count = 0
for value in a_squared_list:
    print(f'the values of the {count}-th record in the list is {value}')
    count += 1
```

```python
print(list(zip(a_list, a_squared_list)))
```

```python
print(list(enumerate(a_squared_list)))
```

```python
# we can use List comprehension to access the list values.
[k for k in enumerate(a_squared_list)]
```

## Open files
Open files behave like lists too! Here we get each line in the file and find its length, using the comprehension syntax to put these lengths into a big list.

```python
#this is a poor way of opening a file. it does not close the file
linelengths = [len(line) for line in open("data/hamlet.txt")]
print(linelengths)
```

```python
sum(linelengths),  min(linelengths), max(linelengths)
```

But perhaps we want to access Hamlet word by word and not line by line

```python
hamletfile = open("data/hamlet.txt")
hamlettext = hamletfile.read()
hamletfile.close()
hamlettokens = hamlettext.split() #split with no arguments splits on whitespace
len(hamlettokens)
```

One can use the `with` syntax which cretaes a context. The file closing is then done automatically for us.

```python
with open("data/hamlet.txt") as hamletfile:
    hamlettext=hamletfile.read()
    hamlettokens=hamlettext.split()
    print(len(hamlettokens))
```

There are roughly 32,000 words in Hamlet.

## Range
`range` gets the list of integers upto N.

```python
mylist = []
for i in range(10):
    mylist.append(i)
mylist
```

```python
mylist = []
for i in range(1,10):
    mylist.append(i)
mylist
```

```python
mylist = []
for i in range(10,0,-1):
    mylist.append(i)
mylist
```

## Functions


```python
def square(x):
    return(x*x)

def cube(x):
    return x*x*x
square(5),cube(5)
```

```python
print(square, type(cube))
```

In Python, functions are "first-class". This is just a fancy way of saying, you can pass functions to other functions

```python
def sum_of_anything(x, y, f):
    print(x,y,f)
    return(f(x) + f(y))
sum_of_anything(3,4,square)
```

Python functions can have positional arguments and keyword arguments. Positional arguments are stored in a tuple, and keyword arguments in a dictionary. Note the "starred" syntax

```python
def f(a, b, *args, **kwargs):
    print("got", a, b, args, kwargs)
    return a
print(f(1,3))
print(f(1, 3, 4, d=1, c=2))
```

## IF
Python supports if/elif/else clauses for multi-way conditionals

```python
def do_it(x):
    if x==1:
        print( "One")
    elif x==2:
        print( "Two")
    else:
        print(x)
do_it(1)
```

```python
do_it(2); do_it(3)
```

You can `break` out of a loop based on a condition. The loop below is a for loop.

```python
for i in range(10):
    print(i)
    if (i > 5):
        break
```

While loops are also supported. `continue` continues to the next iteration of the loop skipping all the code below, while `break` breaks out of it.

```python
i=0
while True:
    print( i)
    i=i+1
    if i < 5:
        continue
    else:
        break
```
